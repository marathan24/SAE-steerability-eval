```python
import torch
from typing import Dict, List, Tuple
from transformer_lens import HookedTransformer
from sae_lens import SAE
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

class INTJSteeringEvaluator:
    def __init__(self, model: HookedTransformer, sae: SAE, max_act: float, steering_feature: int = None):
        """
        Initialize the INTJ Steering Evaluator.

        Args:
            model (HookedTransformer): The transformer model.
            sae (SAE): Sparse autoencoder.
            max_act (float): Maximum activation value for steering.
            steering_feature (int, optional): Specific feature index to steer. Defaults to None.
        """
        self.model = model
        self.sae = sae
        self.max_act = max_act
        self.steering_feature = steering_feature  # Ensure this is set if steering is required

        # Base INTJ statements
        self.agree_statements = [
            "I would analyze a situation from multiple angles before forming an opinion.",
            "I believe in making decisions based on logic and objective data.",
        ]

        self.disagree_statements = [
            "I would rather go with the flow and see where things lead.",
            "I would seek comfort in large gatherings and social events.",
        ]

        self.baseline_activations = None

    def examine_cache_structure(self, text: str):
        """
        Examine the cache structure for a given text.

        Args:
            text (str): The input text.

        Returns:
            Dict: The cache containing activations.
        """
        tokens = self.model.to_tokens(text, prepend_bos=True)
        _, cache = self.model.run_with_cache_with_saes(tokens, saes=[self.sae])
        print("Available cache keys:", [k for k in cache.keys() if 'sae' in k])
        return cache

    def get_activations(
        self,
        text: str,
        steering_strength: float = 0.0
    ) -> torch.Tensor:
        """
        Get feature activations for a given text with optional steering.

        Args:
            text (str): The input text.
            steering_strength (float, optional): Steering strength. Defaults to 0.0.

        Returns:
            torch.Tensor: The activation tensor.
        """
        def steering_hook(acts, hook):
            if self.steering_feature is not None:
                acts[:, :, self.steering_feature] += self.max_act * steering_strength
            return acts

        with self.model.hooks(fwd_hooks=[(self.sae.cfg.hook_name, steering_hook)] if steering_strength != 0 and self.steering_feature is not None else []):
            tokens = self.model.to_tokens(text, prepend_bos=True)
            _, cache = self.model.run_with_cache_with_saes(tokens, saes=[self.sae])

        sae_keys = [k for k in cache.keys() if 'hook_sae_acts_post' in k]
        if not sae_keys:
            raise ValueError("SAE activations not found in cache.")
        sae_key = sae_keys[-1]
        activations = cache[sae_key][0, -1, :]
        return activations

    def establish_baseline(self):
        """
        Establish baseline activations from initial agree/disagree statements.
        """
        baseline_results = []

        for idx, (agree, disagree) in enumerate(zip(self.agree_statements, self.disagree_statements)):
            agree_acts = self.get_activations(agree)
            disagree_acts = self.get_activations(disagree)

            diff = agree_acts - disagree_acts
            baseline_results.append(diff.unsqueeze(0)) 

        self.baseline_activations = torch.cat(baseline_results, dim=0).mean(dim=0)
        print("Baseline activations established.")

    def analyze_new_statement(self, new_statement: str, steering_strength: float = 0.0):
        """
        Analyze a new statement against the baseline.

        Args:
            new_statement (str): The new statement to evaluate.
            steering_strength (float, optional): Steering strength. Defaults to 0.0.

        Returns:
            Dict: Analysis results including feature differences and activation values.
        """
        if self.baseline_activations is None:
            raise ValueError("Baseline activations not established. Call establish_baseline() first.")

        new_activations = self.get_activations(new_statement, steering_strength)

        diff = new_activations - self.baseline_activations
        top_diff_vals, top_diff_idxs = torch.topk(torch.abs(diff), 10)

        analysis = {
            'statement': new_statement,
            'steering_strength': steering_strength,
            'top_diff_features': top_diff_idxs.tolist(),
            'diff_magnitudes': top_diff_vals.tolist(),
            'diff_directions': diff[top_diff_idxs].tolist(),
            'all_diffs': diff.tolist()
        }

        return analysis

    def evaluate_statements(self, test_statements: List[Dict], steering_strength: float = 0.0) -> pd.DataFrame:
        """
        Evaluate multiple test statements against the baseline.

        Args:
            test_statements (List[Dict]): A list of dictionaries with 'text' and 'label' keys.
            steering_strength (float, optional): Steering strength. Defaults to 0.0.

        Returns:
            pd.DataFrame: DataFrame containing analysis results for each statement.
        """
        results = []

        for stmt in test_statements:
            analysis = self.analyze_new_statement(stmt['text'], steering_strength)
            alignment_score = torch.sum(torch.abs(torch.tensor(analysis['diff_directions']))).item()
            results.append({
                'Statement': stmt['text'],
                'Label': stmt['label'],
                'Steering Strength': analysis['steering_strength'],
                'Alignment Score': alignment_score
            })

        df_results = pd.DataFrame(results)
        return df_results

    def visualize_alignment_scores(self, df: pd.DataFrame, title: str = "Persona Steering Alignment Scores"):
        """
        Visualize alignment scores for multiple statements.

        Args:
            df (pd.DataFrame): DataFrame containing alignment scores and labels.
            title (str, optional): Title of the graph. Defaults to "Persona Steering Alignment Scores".
        """
        plt.figure(figsize=(12, 8))
        colors = {'Aligned': 'green', 'Conflicting': 'red'}
        plt.bar(df['Statement'], df['Alignment Score'], color=df['Label'].map(colors), edgecolor='black')

        for idx, row in df.iterrows():
            plt.text(idx, row['Alignment Score'] + 0.05, f"{row['Alignment Score']:.2f}",
                     ha='center', va='bottom', fontsize=8, rotation=90)

        plt.xlabel('Test Statements')
        plt.ylabel('Alignment Score')
        plt.title(title)
        plt.xticks(rotation=90)
        plt.legend(handles=[
            plt.Line2D([0], [0], color='green', lw=4, label='Aligned'),
            plt.Line2D([0], [0], color='red', lw=4, label='Conflicting')
        ])
        plt.tight_layout()
        plt.show()

    def compare_texts(
        self,
        text1: str,
        text2: str,
        steering_strength: float = 0.0
    ) -> Dict:
        """
        Compare activations between two texts.

        Args:
            text1 (str): First text.
            text2 (str): Second text.
            steering_strength (float, optional): Steering strength. Defaults to 0.0.

        Returns:
            Dict: Comparison results with top differing features.
        """
        acts1 = self.get_activations(text1, steering_strength)
        acts2 = self.get_activations(text2, steering_strength)

        diff = acts1 - acts2
        top_diffs, top_idxs = torch.topk(torch.abs(diff), 10)

        comparison = {
            'text1': text1,
            'text2': text2,
            'steering_strength': steering_strength,
            'top_diff_features': top_idxs.tolist(),
            'diff_magnitudes': top_diffs.tolist(),
            'diff_directions': diff[top_idxs].tolist()
        }

        return comparison

    def visualize_comparison(
        self,
        comparison_results: Dict,
        title: str = "Text Comparison Activation Differences"
    ):
        """
        Visualize differences between two texts using Matplotlib.

        Args:
            comparison_results (Dict): The comparison results containing feature differences.
            title (str, optional): Title of the graph. Defaults to "Text Comparison Activation Differences".
        """
        features = comparison_results['top_diff_features']
        magnitudes = comparison_results['diff_magnitudes']
        directions = comparison_results['diff_directions']

        colors = ['blue' if direction > 0 else 'orange' for direction in directions]

        plt.figure(figsize=(10, 6))
        scatter = plt.scatter(features, magnitudes, c=colors, s=100, edgecolors='black')

        for i, (feature, magnitude, direction) in enumerate(zip(features, magnitudes, directions)):
            plt.text(feature, magnitude, f'{direction:.2f}', fontsize=9,
                     ha='center', va='bottom')

        plt.xlabel('Feature Index')
        plt.ylabel('Difference Magnitude')
        plt.title(title)
        plt.xticks(features)
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.tight_layout()
        plt.show()

    def visualize_activations(
        self,
        analysis_results: Dict,
        title: str = "Activation Differences"
    ):
        """
        Visualize activation differences using Matplotlib.

        Args:
            analysis_results (Dict): The analysis results containing feature differences.
            title (str, optional): Title of the graph. Defaults to "Activation Differences".
        """
        features = analysis_results['top_diff_features']
        magnitudes = analysis_results['diff_magnitudes']
        directions = analysis_results['diff_directions']

        # Determine colors based on direction
        colors = ['green' if direction > 0 else 'red' for direction in directions]

        plt.figure(figsize=(10, 6))
        bars = plt.bar(features, magnitudes, color=colors, edgecolor='black')

        # Add annotations
        for bar, direction in zip(bars, directions):
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{direction:.2f}',
                     ha='center', va='bottom', fontsize=8, color='black')

        plt.xlabel('Feature Index')
        plt.ylabel('Difference Magnitude')
        plt.title(title)
        plt.xticks(features)
        plt.tight_layout()
        plt.show()

    def summarize_alignment(self, df: pd.DataFrame):
        """
        Print summary statistics for alignment scores.

        Args:
            df (pd.DataFrame): DataFrame containing alignment scores.
        """
        summary = df.groupby('Label')['Alignment Score'].describe()
        print("Alignment Score Summary by Label:")
        print(summary)


steering_feature_index = 9264 # NO need to use this until we know which is the feature we are concerned about the most
max_activation = 183.1667

evaluator = INTJSteeringEvaluator(model, sae, max_act=max_activation)

evaluator.establish_baseline()

test_statements = [
    # Statements an INTJ would typically align with
    {"text": "I would analyze a situation from multiple angles before forming an opinion.", "label": "Aligned"},
    {"text": "I believe in making decisions based on logic and objective data.", "label": "Aligned"},
    {"text": "I would prioritize understanding the underlying principles behind a concept before implementing it.", "label": "Aligned"},
    {"text": "I prefer to delve into complex subjects that require critical thinking.", "label": "Aligned"},
    {"text": "I would create detailed plans and timelines for projects to ensure efficient execution.", "label": "Aligned"},
    {"text": "I believe in identifying patterns and connections that others might miss.", "label": "Aligned"},
    {"text": "I would prefer to work independently rather than in a highly collaborative environment.", "label": "Aligned"},
    {"text": "I would choose to recharge by spending time alone reflecting and analyzing my thoughts rather than socializing.", "label": "Aligned"},
    {"text": "I believe that understanding the root causes of a problem is essential for finding effective solutions.", "label": "Aligned"},
    {"text": "I would prefer to communicate my ideas through concise and well-structured writing.", "label": "Aligned"},
    {"text": "I like to work independently rather than as part of a team.", "label": "Aligned"},

    
    
    # Statements an INTJ would typically find conflicting
    {"text": "I would rather go with the flow and see where things lead.", "label": "Conflicting"},
    {"text": "I would prioritize building strong connections with others over achieving my goals.", "label": "Conflicting"},
    {"text": "I would seek comfort in large gatherings and social events.", "label": "Conflicting"},
    {"text": "I would rather listen to my heart than my head.", "label": "Conflicting"},
    {"text": "I believe that small talk and casual conversations are important for building relationships.", "label": "Conflicting"},
    {"text": "I would trust my gut feeling over a carefully researched analysis.", "label": "Conflicting"},
    {"text": "I would make decisions based on how they make me feel.", "label": "Conflicting"},
    {"text": "I believe that it's more important to be well-liked than to be right.", "label": "Conflicting"},
    {"text": "I prefer spontaneous activities over structured plans.", "label": "Conflicting"},

    {"text": "I would be more motivated by praise and encouragement than by recognition for my accomplishments.", "label": "Conflicting"},
    {"text": "I believe it's important to be spontaneous and flexible, even if it means deviating from plans.", "label": "Conflicting"}
]

alignment_df = evaluator.evaluate_statements(test_statements, steering_strength=1.0)  # Adjust steering_strength as needed
print("\nAlignment DataFrame:")
print(alignment_df)

evaluator.visualize_alignment_scores(alignment_df, title="Persona Steering Alignment Scores for INTJ")

evaluator.summarize_alignment(alignment_df)

with open('analysis_results.txt', 'a') as f:
    for stmt in test_statements:
        analysis = evaluator.analyze_new_statement(stmt['text'], steering_strength=1)
        
        print(f"\nAnalysis for Statement: {stmt['text']}")
        print("Top Differing Features:", analysis['top_diff_features'])
        print("Difference Magnitudes:", analysis['diff_magnitudes'])
        print("Difference Directions:", analysis['diff_directions'])
        
        f.write(f"\nAnalysis for Statement: {stmt['text']}\n")
        f.write(f"Top Differing Features: {analysis['top_diff_features']}\n")
        f.write(f"Difference Magnitudes: {analysis['diff_magnitudes']}\n")
        f.write(f"Difference Directions: {analysis['diff_directions']}\n")
        f.flush()  
        
        evaluator.visualize_activations(analysis, title=f"Activation Differences for: \"{stmt['text']}\"")
```

```python
import torch
from typing import Dict, List, Tuple
from transformer_lens import HookedTransformer
from sae_lens import SAE
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

class INTJSteeringEvaluator:
    def __init__(self, model: HookedTransformer, sae: SAE, max_act: float, steering_feature: int = None):
        """
        Initialize the INTJ Steering Evaluator.

        Args:
            model (HookedTransformer): The transformer model.
            sae (SAE): Sparse autoencoder.
            max_act (float): Maximum activation value for steering.
            steering_feature (int, optional): Specific feature index to steer. Defaults to None.
        """
        self.model = model
        self.sae = sae
        self.max_act = max_act
        self.steering_feature = steering_feature  

        self.agree_statements = [
            "I would analyze a situation from multiple angles before forming an opinion.",
            "I believe in making decisions based on logic and objective data.",
            
        ]

        self.disagree_statements = [
            "I would rather go with the flow and see where things lead.",
            "I would seek comfort in large gatherings and social events.",
        ]

        self.baseline_activations = None

    def create_baseline_prompt(self) -> str:
        """
        Create the prompt for establishing baseline activations.

        Returns:
            str: The baseline prompt.
        """
        return f"""You are role playing as a persona described as follows: INTJ

The following are statements that this persona agrees with:
{chr(10).join(['- ' + s for s in self.agree_statements])}

The following are statements that other people would agree with, but this persona would disagree with:
{chr(10).join(['- ' + s for s in self.disagree_statements])}"""

    def create_analysis_prompt(self, statement: str) -> str:
        """
        Create the prompt for analyzing a new statement.

        Args:
            statement (str): The new statement to evaluate.

        Returns:
            str: The analysis prompt.
        """
        return f"""You are role playing as a persona described as follows: INTJ

The following are statements that this persona agrees with:
{chr(10).join(['- ' + s for s in self.agree_statements])}

The following are statements that other people would agree with, but this persona would disagree with:
{chr(10).join(['- ' + s for s in self.disagree_statements])}

You will now be given a new statement. Your job is to determine, based on your understanding of the persona, whether they would agree or disagree with it.

Response should be in valid JSON with:
{{"agree": true}} if the persona would agree, or {{"agree": false}} if they would disagree.

Statement: {statement}
Response:"""

    def examine_cache_structure(self, text: str):
        """
        Examine the cache structure for a given text.

        Args:
            text (str): The input text.

        Returns:
            Dict: The cache containing activations.
        """
        tokens = self.model.to_tokens(text, prepend_bos=True)
        _, cache = self.model.run_with_cache_with_saes(tokens, saes=[self.sae])
        print("Available cache keys:", [k for k in cache.keys() if 'sae' in k])
        return cache

    def get_activations(
        self,
        text: str,
        steering_strength: float = 0.0
    ) -> torch.Tensor:
        """
        Get feature activations for a given text with optional steering.

        Args:
            text (str): The input text.
            steering_strength (float, optional): Steering strength. Defaults to 0.0.

        Returns:
            torch.Tensor: The activation tensor.
        """
        def steering_hook(acts, hook):
            if self.steering_feature is not None:
                acts[:, :, self.steering_feature] += self.max_act * steering_strength
            return acts

        with self.model.hooks(fwd_hooks=[(self.sae.cfg.hook_name, steering_hook)] if steering_strength != 0 and self.steering_feature is not None else []):
            tokens = self.model.to_tokens(text, prepend_bos=True)
            _, cache = self.model.run_with_cache_with_saes(tokens, saes=[self.sae])

        sae_keys = [k for k in cache.keys() if 'hook_sae_acts_post' in k]
        if not sae_keys:
            raise ValueError("SAE activations not found in cache.")
        sae_key = sae_keys[-1]  
        activations = cache[sae_key][0, -1, :]
        return activations

    def establish_baseline(self):
        """
        Establish baseline activations from the combined agree and disagree statements.
        """
        prompt = self.create_baseline_prompt()
        baseline_acts = self.get_activations(prompt)
        self.baseline_activations = baseline_acts
        print("Baseline activations established.")

    def analyze_new_statement(self, new_statement: str, steering_strength: float = 0.0):
        """
        Analyze a new statement against the baseline.

        Args:
            new_statement (str): The new statement to evaluate.
            steering_strength (float, optional): Steering strength. Defaults to 0.0.

        Returns:
            Dict: Analysis results including feature differences and activation values.
        """
        if self.baseline_activations is None:
            raise ValueError("Baseline activations not established. Call establish_baseline() first.")

        prompt = self.create_analysis_prompt(new_statement)
        new_activations = self.get_activations(prompt, steering_strength)

        diff = new_activations - self.baseline_activations

        top_diff_vals, top_diff_idxs = torch.topk(torch.abs(diff), 10)

        analysis = {
            'statement': new_statement,
            'steering_strength': steering_strength,
            'top_diff_features': top_diff_idxs.tolist(),
            'diff_magnitudes': top_diff_vals.tolist(),
            'diff_directions': diff[top_diff_idxs].tolist(),
            'all_diffs': diff.tolist()
        }

        return analysis

    def evaluate_statements(self, test_statements: List[Dict], steering_strength: float = 0.0) -> pd.DataFrame:
        """
        Evaluate multiple test statements against the baseline.

        Args:
            test_statements (List[Dict]): A list of dictionaries with 'text' and 'label' keys.
            steering_strength (float, optional): Steering strength. Defaults to 0.0.

        Returns:
            pd.DataFrame: DataFrame containing analysis results for each statement.
        """
        results = []

        for stmt in test_statements:
            analysis = self.analyze_new_statement(stmt['text'], steering_strength)
            alignment_score = torch.sum(torch.abs(torch.tensor(analysis['diff_directions']))).item()
            results.append({
                'Statement': stmt['text'],
                'Label': stmt['label'],  
                'Steering Strength': analysis['steering_strength'],
                'Alignment Score': alignment_score
            })

        df_results = pd.DataFrame(results)
        return df_results

    def visualize_alignment_scores(self, df: pd.DataFrame, title: str = "Persona Steering Alignment Scores"):
        """
        Visualize alignment scores for multiple statements.

        Args:
            df (pd.DataFrame): DataFrame containing alignment scores and labels.
            title (str, optional): Title of the graph. Defaults to "Persona Steering Alignment Scores".
        """
        plt.figure(figsize=(12, 8))
        colors = {'Aligned': 'green', 'Conflicting': 'red'}
        plt.bar(df['Statement'], df['Alignment Score'], color=df['Label'].map(colors), edgecolor='black')

        for idx, row in df.iterrows():
            plt.text(idx, row['Alignment Score'] + 0.05, f"{row['Alignment Score']:.2f}",
                     ha='center', va='bottom', fontsize=8, rotation=90)

        plt.xlabel('Test Statements')
        plt.ylabel('Alignment Score')
        plt.title(title)
        plt.xticks(rotation=90)
        plt.legend(handles=[
            plt.Line2D([0], [0], color='green', lw=4, label='Aligned'),
            plt.Line2D([0], [0], color='red', lw=4, label='Conflicting')
        ])
        plt.tight_layout()
        plt.show()

    def compare_texts(
        self,
        text1: str,
        text2: str,
        steering_strength: float = 0.0
    ) -> Dict:
        """
        Compare activations between two texts.

        Args:
            text1 (str): First text.
            text2 (str): Second text.
            steering_strength (float, optional): Steering strength. Defaults to 0.0.

        Returns:
            Dict: Comparison results with top differing features.
        """
        acts1 = self.get_activations(text1, steering_strength)
        acts2 = self.get_activations(text2, steering_strength)

        diff = acts1 - acts2
        top_diffs, top_idxs = torch.topk(torch.abs(diff), 10)

        comparison = {
            'text1': text1,
            'text2': text2,
            'steering_strength': steering_strength,
            'top_diff_features': top_idxs.tolist(),
            'diff_magnitudes': top_diffs.tolist(),
            'diff_directions': diff[top_idxs].tolist()
        }

        return comparison

    def visualize_comparison(
        self,
        comparison_results: Dict,
        title: str = "Text Comparison Activation Differences"
    ):
        """
        Visualize differences between two texts using Matplotlib.

        Args:
            comparison_results (Dict): The comparison results containing feature differences.
            title (str, optional): Title of the graph. Defaults to "Text Comparison Activation Differences".
        """
        features = comparison_results['top_diff_features']
        magnitudes = comparison_results['diff_magnitudes']
        directions = comparison_results['diff_directions']

        colors = ['blue' if direction > 0 else 'orange' for direction in directions]

        plt.figure(figsize=(10, 6))

        for i, (feature, magnitude, direction) in enumerate(zip(features, magnitudes, directions)):
            plt.text(feature, magnitude, f'{direction:.2f}', fontsize=9,
                     ha='center', va='bottom')

        plt.xlabel('Feature Index')
        plt.ylabel('Difference Magnitude')
        plt.title(title)
        plt.xticks(features)
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.tight_layout()
        plt.show()

    def visualize_activations(
        self,
        analysis_results: Dict,
        title: str = "Activation Differences"
    ):
        """
        Visualize activation differences using Matplotlib.

        Args:
            analysis_results (Dict): The analysis results containing feature differences.
            title (str, optional): Title of the graph. Defaults to "Activation Differences".
        """
        features = analysis_results['top_diff_features']
        magnitudes = analysis_results['diff_magnitudes']
        directions = analysis_results['diff_directions']

        colors = ['green' if direction > 0 else 'red' for direction in directions]

        plt.figure(figsize=(10, 6))
        bars = plt.bar(features, magnitudes, color=colors, edgecolor='black')

        for bar, direction in zip(bars, directions):
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{direction:.2f}',
                     ha='center', va='bottom', fontsize=8, color='black')

        plt.xlabel('Feature Index')
        plt.ylabel('Difference Magnitude')
        plt.title(title)
        plt.xticks(features)
        plt.tight_layout()
        plt.show()

    def summarize_alignment(self, df: pd.DataFrame):
        """
        Print summary statistics for alignment scores.

        Args:
            df (pd.DataFrame): DataFrame containing alignment scores.
        """
        summary = df.groupby('Label')['Alignment Score'].describe()
        print("Alignment Score Summary by Label:")
        print(summary)




steering_feature_index = 9264  
max_activation = 183.1667

evaluator = INTJSteeringEvaluator(model, sae, max_act=max_activation)

evaluator.establish_baseline()

test_statements = [
    # Statements an INTJ would typically align with
    {"text": "I would analyze a situation from multiple angles before forming an opinion.", "label": "Aligned"},
    {"text": "I believe in making decisions based on logic and objective data.", "label": "Aligned"},
    {"text": "I would prioritize understanding the underlying principles behind a concept before implementing it.", "label": "Aligned"},
    {"text": "I prefer to delve into complex subjects that require critical thinking.", "label": "Aligned"},
    {"text": "I would create detailed plans and timelines for projects to ensure efficient execution.", "label": "Aligned"},
    {"text": "I believe in identifying patterns and connections that others might miss.", "label": "Aligned"},
    {"text": "I would prefer to work independently rather than in a highly collaborative environment.", "label": "Aligned"},
    {"text": "I would choose to recharge by spending time alone reflecting and analyzing my thoughts rather than socializing.", "label": "Aligned"},
    {"text": "I believe that understanding the root causes of a problem is essential for finding effective solutions.", "label": "Aligned"},
    {"text": "I would prefer to communicate my ideas through concise and well-structured writing.", "label": "Aligned"},
    {"text": "I like to work independently rather than as part of a team.", "label": "Aligned"},
    {"text": "I prefer to work independently", "label":"Aligned"},
    
    



    # Statements an INTJ would typically find conflicting
    {"text": "I would rather go with the flow and see where things lead.", "label": "Conflicting"},
    {"text": "I would prioritize building strong connections with others over achieving my goals.", "label": "Conflicting"},
    {"text": "I would seek comfort in large gatherings and social events.", "label": "Conflicting"},
    {"text": "I would rather listen to my heart than my head.", "label": "Conflicting"},
    {"text": "I believe that small talk and casual conversations are important for building relationships.", "label": "Conflicting"},
    {"text": "I would trust my gut feeling over a carefully researched analysis.", "label": "Conflicting"},
    {"text": "I would make decisions based on how they make me feel.", "label": "Conflicting"},
    {"text": "I believe that it's more important to be well-liked than to be right.", "label": "Conflicting"},
    {"text": "I prefer spontaneous activities over structured plans.", "label": "Conflicting"},

    {"text": "I would be more motivated by praise and encouragement than by recognition for my accomplishments.", "label": "Conflicting"},
    {"text": "I believe it's important to be spontaneous and flexible, even if it means deviating from plans.", "label": "Conflicting"}
]

alignment_df = evaluator.evaluate_statements(test_statements, steering_strength=1.0)  
print("\nAlignment DataFrame:")
print(alignment_df)

evaluator.visualize_alignment_scores(alignment_df, title="Persona Steering Alignment Scores for INTJ")

evaluator.summarize_alignment(alignment_df)


with open('analysis_results_few_shot_applied.txt', 'a') as f:
    for stmt in test_statements:
        analysis = evaluator.analyze_new_statement(stmt['text'], steering_strength=1.0)
        
        print(f"\nAnalysis for Statement: {stmt['text']}")
        print("Top Differing Features:", analysis['top_diff_features'])
        print("Difference Magnitudes:", analysis['diff_magnitudes'])
        print("Difference Directions:", analysis['diff_directions'])
        
        f.write(f"\nAnalysis for Statement: {stmt['text']}\n")
        f.write(f"Top Differing Features: {analysis['top_diff_features']}\n")
        f.write(f"Difference Magnitudes: {analysis['diff_magnitudes']}\n")
        f.write(f"Difference Directions: {analysis['diff_directions']}\n")
        f.flush()
        
        evaluator.visualize_activations(analysis, title=f"Activation Differences for: \"{stmt['text']}\"")
```

# CLEAN MEMORY

```python
import torch
import gc
import sys

def clear_gpu_memory():
    """
    Clears GPU memory by deleting all CUDA tensors and models, running garbage collection, and emptying the CUDA cache.
    """
    for obj in gc.get_objects():
        try:
            if torch.is_tensor(obj) and obj.is_cuda:
                del obj
            elif hasattr(obj, 'data') and torch.is_tensor(obj.data) and obj.data.is_cuda:
                del obj.data
        except Exception as e:
            pass

    gc.collect()

    torch.cuda.empty_cache()

    print(f"GPU memory allocated: {torch.cuda.memory_allocated(0) / 1024**2:.2f} MB")
    print(f"GPU memory reserved: {torch.cuda.memory_reserved(0) / 1024**2:.2f} MB")

def delete_all_models_and_tensors():
    """
    Deletes all variables in the global namespace to remove references 
    to models and tensors.
    """
    global_vars = list(globals().keys())
    
    for var in global_vars:
        obj = globals()[var]
        try:
            if isinstance(obj, torch.nn.Module):
                del globals()[var]
            elif torch.is_tensor(obj) and obj.is_cuda:
                del globals()[var]
        except Exception:
            pass

    local_vars = list(locals().keys())
    for var in local_vars:
        obj = locals()[var]
        try:
            if isinstance(obj, torch.nn.Module):
                del locals()[var]
            elif torch.is_tensor(obj) and obj.is_cuda:
                del locals()[var]
        except Exception:
            pass

def reset_gpu_memory():
    """
    High-level function to delete models and tensors, then clear GPU memory.
    """
    delete_all_models_and_tensors()
    clear_gpu_memory()
    torch.cuda.reset_peak_memory_stats()


reset_gpu_memory()

print("After clearing:")
print(f"GPU memory allocated: {torch.cuda.memory_allocated(0) / 1024**2:.2f} MB")
print(f"GPU memory reserved: {torch.cuda.memory_reserved(0) / 1024**2:.2f} MB")
```
